```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def plot_simplex_geometry():
    # --- 1. Setup the Geometry (Equilateral Triangle) ---
    # Vertices of the simplex in 2D Cartesian coordinates
    # A (Top), B (Left), C (Right)
    # This maps probability vectors [p1, p2, p3] to 2D points
    vertices = np.array([[0.5, np.sqrt(3)/2], [0, 0], [1, 0]])
    labels = ['Outcome 1', 'Outcome 2', 'Outcome 3']

    def to_cartesian(p):
        """Maps a probability distribution p (shape K) to 2D simplex coordinates."""
        return p @ vertices

    # --- 2. Define the Credal Set (The "Bad" Failure Modes) ---
    # We define 3 "Bad" distributions that form the hull
    bad_modes = np.array([
        [0.40, 0.40, 0.20], # Mixed failure
        [0.10, 0.80, 0.10], # Mode collapse to 2
        [0.10, 0.10, 0.80]  # Mode collapse to 3
    ])
    
    # --- 3. Define the User (The "Good" Agent) ---
    # A distribution safely outside the bad hull
    Q_user = np.array([0.80, 0.10, 0.10]) # Strong performance on Outcome 1

    # --- 4. Solve for P* (Reverse Info Projection) ---
    # Find P in ConvexHull(bad_modes) that minimizes KL(Q || P)
    # P is a convex combination of bad_modes: P = w @ bad_modes
    
    def objective(w):
        # w are weights on the bad modes
        P_mix = w @ bad_modes
        # KL Divergence: sum Q * log(Q/P)
        # We minimize KL(Q||P) -> Maximize sum Q * log(P)
        return -np.sum(Q_user * np.log(P_mix + 1e-9))

    cons = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})
    bnds = tuple((0, 1) for _ in range(len(bad_modes)))
    
    # Optimize weights
    res = minimize(objective, [1/3]*3, bounds=bnds, constraints=cons)
    w_star = res.x
    P_star = w_star @ bad_modes
    
    # --- 5. Construct the Separating Hyperplane (The Martingale) ---
    # The optimal payoff S = Q / P_star
    # The decision boundary is where E_P[S] = 1
    # Equation: p1*S1 + p2*S2 + p3*S3 = 1
    # In barycentric: p1 + p2 + p3 = 1
    # We need to plot this line in 2D. 
    # We find two points on the edges of the simplex where this equation holds.
    S = Q_user / P_star
    
    # --- PLOTTING ---
    plt.figure(figsize=(10, 9))
    
    # A. Draw the Simplex Border
    triangle = np.vstack([vertices, vertices[0]])
    plt.plot(triangle[:,0], triangle[:,1], 'k-', lw=1, alpha=0.3)
    
    # B. Draw the Credal Set (Convex Hull of Bad Modes)
    # We sample points in the hull to fill it or just plot the polygon
    hull_coords = bad_modes @ vertices
    # Close the loop for plotting
    hull_plot = np.vstack([hull_coords, hull_coords[0]])
    
    plt.fill(hull_plot[:,0], hull_plot[:,1], color='gray', alpha=0.3, label='Credal Set (Null Hypothesis)')
    plt.plot(hull_plot[:,0], hull_plot[:,1], 'k--', lw=1)
    
    # C. Plot Points
    coord_Q = to_cartesian(Q_user)
    coord_P_star = to_cartesian(P_star)
    
    plt.scatter(coord_Q[0], coord_Q[1], color='blue', s=150, zorder=5, label='User Q (Good)')
    plt.scatter(coord_P_star[0], coord_P_star[1], color='red', s=150, zorder=5, label='Projection P* (Least Favorable)')
    
    # D. Plot the Separating Hyperplane (The Martingale Line)
    # We iterate over edges to find intersection points of p.dot(S) = 1
    # Edge 1: p3 = 0 -> p1*S1 + (1-p1)*S2 = 1
    intersections = []
    
    # Helper to check if point is valid probability
    def is_valid(p): return np.all(p >= -1e-5) and np.all(p <= 1.00001)

    # Solve for edges (p_i, p_j varying, p_k=0)
    # Edge 1-2 (p3=0): p1*S1 + (1-p1)*S2 = 1 => p1(S1-S2) = 1 - S2
    if abs(S[0]-S[1]) > 1e-5:
        p1 = (1 - S[1]) / (S[0] - S[1])
        p = np.array([p1, 1-p1, 0])
        if is_valid(p): intersections.append(to_cartesian(p))
            
    # Edge 2-3 (p1=0): p2*S2 + (1-p2)*S3 = 1
    if abs(S[1]-S[2]) > 1e-5:
        p2 = (1 - S[2]) / (S[1] - S[2])
        p = np.array([0, p2, 1-p2])
        if is_valid(p): intersections.append(to_cartesian(p))
            
    # Edge 3-1 (p2=0): p3*S3 + (1-p3)*S1 = 1
    if abs(S[2]-S[0]) > 1e-5:
        p3 = (1 - S[0]) / (S[2] - S[0])
        p = np.array([1-p3, 0, p3])
        if is_valid(p): intersections.append(to_cartesian(p))

    if len(intersections) >= 2:
        intersections = np.array(intersections)
        # Extend the line a bit for visual effect
        plt.plot(intersections[:,0], intersections[:,1], color='green', lw=3, label="Breakeven Line (E[S]=1)")

    # E. Draw Influence Vector (Arrow from P* to Q)
    plt.arrow(coord_P_star[0], coord_P_star[1], 
              coord_Q[0]-coord_P_star[0], coord_Q[1]-coord_P_star[1], 
              color='black', width=0.005, length_includes_head=True)
    plt.text((coord_Q[0]+coord_P_star[0])/2 + 0.02, (coord_Q[1]+coord_P_star[1])/2, 
             r'$\psi_Q$ (Influence)', fontsize=12)

    # Labels
    plt.text(vertices[0,0], vertices[0,1]+0.05, 'Outcome 1', ha='center')
    plt.text(vertices[1,0]-0.1, vertices[1,1]-0.05, 'Outcome 2', ha='center')
    plt.text(vertices[2,0]+0.1, vertices[2,1]-0.05, 'Outcome 3', ha='center')

    plt.legend(loc='upper right')
    plt.title("Visual Proof: Convex Separation of Credal Sets", fontsize=14)
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    plot_simplex_geometry()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_non_convex_failure():
    # --- 1. Setup Geometry ---
    vertices = np.array([[0.5, np.sqrt(3)/2], [0, 0], [1, 0]])
    
    def to_cartesian(p):
        return p @ vertices

    # --- 2. Define Non-Convex Bad Set (Two Disjoint Islands) ---
    # Island 1: Concentrated near Outcome 2 (Bottom Left)
    # Island 2: Concentrated near Outcome 3 (Bottom Right)
    
    # We generate points for visual "blobs"
    n_points = 100
    noise = 0.05
    
    # Blob 1 (Near [0, 1, 0])
    blob1 = []
    for _ in range(n_points):
        p = np.array([0.1, 0.8, 0.1]) + np.random.normal(0, noise, 3)
        p = np.abs(p) / np.sum(np.abs(p))
        blob1.append(to_cartesian(p))
    blob1 = np.array(blob1)

    # Blob 2 (Near [0, 0, 1])
    blob2 = []
    for _ in range(n_points):
        p = np.array([0.1, 0.1, 0.8]) + np.random.normal(0, noise, 3)
        p = np.abs(p) / np.sum(np.abs(p))
        blob2.append(to_cartesian(p))
    blob2 = np.array(blob2)

    # --- 3. Define the User (Q) ---
    # The User is a 50/50 mix of the two bad modes (High Variance)
    # Q = [0.1, 0.45, 0.45] roughly
    Q_user = np.array([0.1, 0.45, 0.45])
    coord_Q = to_cartesian(Q_user)

    # --- PLOTTING ---
    plt.figure(figsize=(10, 9))
    
    # A. Draw Simplex
    triangle = np.vstack([vertices, vertices[0]])
    plt.plot(triangle[:,0], triangle[:,1], 'k-', lw=1, alpha=0.3)
    
    # B. Plot the Non-Convex Bad Set
    plt.scatter(blob1[:,0], blob1[:,1], color='gray', alpha=0.5, label='Bad Set (Part 1)')
    plt.scatter(blob2[:,0], blob2[:,1], color='gray', alpha=0.5, label='Bad Set (Part 2)')
    
    # C. Plot User
    plt.scatter(coord_Q[0], coord_Q[1], color='blue', s=200, zorder=5, label='User Q (Safe Mix)')

    # D. Visualize the Convex Hull Trap
    # Draw a line connecting the centers of the two blobs
    center1 = np.mean(blob1, axis=0)
    center2 = np.mean(blob2, axis=0)
    plt.plot([center1[0], center2[0]], [center1[1], center2[1]], 'k--', alpha=0.4, 
             label='Convex Hull (Implicit)')

    # E. Draw "Failed" Separating Lines
    # Attempt 1: Separate User from Blob 1
    # Line equation roughly vertical
    plt.plot([0.4, 0.4], [0, 0.3], 'r--', lw=2, label='Failed Linear Test 1')
    
    # Attempt 2: Separate User from Blob 2
    plt.plot([0.6, 0.6], [0, 0.3], 'r--', lw=2, label='Failed Linear Test 2')

    # Attempt 3: Horizontal line (Trying to separate from bottom)
    # This fails because User is at same "height" (same p1) as bad modes
    plt.plot([0.1, 0.9], [0.15, 0.15], 'r--', lw=2, label='Failed Linear Test 3')

    # Annotations
    plt.text(center1[0], center1[1]+0.05, "Bad Mode A", ha='center', fontsize=12)
    plt.text(center2[0], center2[1]+0.05, "Bad Mode B", ha='center', fontsize=12)
    plt.text(coord_Q[0], coord_Q[1]+0.05, "User Q\n(Trapped)", ha='center', color='blue', fontsize=12, fontweight='bold')

    plt.text(vertices[0,0], vertices[0,1]+0.05, 'Outcome 1', ha='center')
    plt.text(vertices[1,0]-0.1, vertices[1,1]-0.05, 'Outcome 2', ha='center')
    plt.text(vertices[2,0]+0.1, vertices[2,1]-0.05, 'Outcome 3', ha='center')

    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    plot_non_convex_failure()
```
